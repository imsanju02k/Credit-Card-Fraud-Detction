```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
data=pd.read_csv("creditcard.csv")
```

```python
# this dataset is highly unblanced

# 0- NOrmal transcation
# 1- fraud transcation
```

```python
# seprating the data for analysis
# series data type in pandas

legit=data[data.Class == 0]
fraud=data[data.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
# statistical measure of the data

legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
# compare the values for both transactions

data.groupby("Class").mean()
```

```python
# dealing with unblanced data
# build the sample data set containing similar distribution of the normal tracnsations and fraudulent transactions
# number of fraudulent transactions =492
# will we combine both the data set and make balanced data set for our analysis

######## using random sampleing ########
```

```python
legit_sample=legit.sample(n=492)
```

```python
# concatenating two data frames
```

```python
new_dataset=pd.concat([legit_sample,fraud],axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset["Class"].value_counts()
```

```python
new_dataset.groupby("Class").mean()
```

```python
# splitting the data into features and targets
```

```python
X=new_dataset.drop(columns="Class",axis=1)
Y=new_dataset["Class"]
```

```python
print(X)
```

```python
print(Y)
```

```python
# split the data into trainning data and testing data
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

```python
# Model Training
#logistical Regression
```

```python
model=LogisticRegression()
```

```python
# training the logistic regression model with training data

model.fit(X_train,Y_train)
```

```python
#Model Evaluation
# Accuracy Score
```

```python
# accuracy on trainning data

X_train_prediction=model.predict(X_train)
training_data_accuracy=accuracy_score(X_train_prediction,Y_train)
```

```python
print ("Accuracy on Training data :",training_data_accuracy)
```

```python
# test data accuracy
X_test_prediction=model.predict(X_test)
test_data_accuracy=accuracy_score(X_test_prediction,Y_test)
```

```python
print("Accuracy on Test Data :",test_data_accuracy)
```
